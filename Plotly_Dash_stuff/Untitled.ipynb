```python
import os
import dash
from dash.dependencies import Input, Output
import dash_daq as daq
import dash_core_components as dcc
import dash_html_components as html

def get_last_line():
    with open("/Users/Tennyson/temps_outfile.csv", "rb") as f:
        try:  # catch OSError in case of a one line file 
            f.seek(-2, os.SEEK_END)
            while f.read(1) != b'\n':
                f.seek(-2, os.SEEK_CUR)
        except OSError:
            f.seek(0)
        last_line = f.readline().decode()
        return last_line

initVal = get_last_line()

app = dash.Dash(__name__)

app.layout = html.Div([
    daq.LEDDisplay(
        id='my-thermometer-1',
        label="Temperature",
        value=int(initVal.split(",")[0]),
        size=256,
        color='black',
        backgroundColor="#777777"
    ),
    dcc.Interval(
            id='interval-component',
            interval=1*1000, # in milliseconds
            n_intervals=0
        )
])

@app.callback(
    Output('my-thermometer-1', 'value'),
    Input('interval-component', 'n_intervals')
)
def update_thermometer(n):
    value=get_last_line()
    return int(value.strip().split(",")[0])

@app.callback(
    Output('my-thermometer-1', 'color'),
    [Input('interval-component', 'n_intervals')]
)
def update_temp_col(n):
    heatVal=get_last_line()
    if heatVal.strip().split(",")[2] == "1":
        return 'orange'
    else:
        return 'black'

@app.callback(
    Output('my-thermometer-1', 'backgroundColor'),
    [Input('interval-component', 'n_intervals')]
)
def update_bg_col(n):
    heatVal=get_last_line()
    if heatVal.strip().split(",")[2] == "1":
        return '#FF5E5E'
    else:
        return '#777777'

@app.callback(
    Output('my-thermometer-1', 'label'),
    [Input('interval-component', 'n_intervals')]
)
def update_label(n):
    heatVal=get_last_line()
    if heatVal.strip().split(",")[2] == "1":
        return f'Heating to {heatVal.strip().split(",")[1]}'
    else:
        return 'Temperature'

if __name__ == '__main__':
    app.run_server(debug=False)
```
